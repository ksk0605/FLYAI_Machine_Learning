```python
def store(train, test, i):
    X_train = train[train.Store == i].iloc[:,:-1].drop(columns='Date')
    y_train = train[train.Store == i].iloc[:,-1]
    test = test[test.Store == i].drop(columns='Date')
    
    train = train.reset_index()

    return X_train,y_train, test
```

```python
from lightgbm import LGBMRegressor
```

```python
def model(X_train, y_train, xtest):
    lgbm = LGBMRegressor()
    lgbm.fit(X_train, y_train, eval_metric='auc', verbose=True)
    pred = lgbm.predict(xtest)
    return pred
```

```python
from sklearn.preprocessing import MinMaxScaler

def wjdrb(xtrain, xtest):
    scaler = MinMaxScaler()
    xtrain = pd.DataFrame(scaler.fit_transform(xtrain))
    xtest = pd.DataFrame(scaler.fit_transform(xtest))
    return xtrain, xtest
```

```python
result = []
for i in range(1, 46):
    X_train,y_train, xtest= store(train, test, i)
    #X_train, xtest = wjdrb(X_train, xtest)
    predict = model(X_train, y_train, xtest)
    result.append(predict)
```
