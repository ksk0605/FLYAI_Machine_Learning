# 여러개의 Plot생성

## 1번째 내부적 생성

```python
plt.figure()#내부적으로 공간만 만들어진 상태
plt.subplot(2,2, 1)#밑으로 한개 옆으로 두개를 그릴 것인데 
plt.plot(xarr)
plt.subplot(2,2, 2)# 각각 순서를 지정가능
plt.plot(xarr)
plt.show()
```

```python
plt.figure()#내부적으로 공간만 만들어진 상태
plt.subplot(2,2, 1)#밑으로 한개 옆으로 두개를 그릴 것인데 

plt.subplot(2,2, 2)# 각각 순서를 지정가능
plt.plot(xarr)
plt.plot(yarr) # 순서가 중요
plt.show()
```

## 2번째 명시적 작성

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax1.plot(xarr)
ax2 = fig.add_subplot(2,2,2)
ax2.plot(xarr)
#object로 받은 것이기 때문에 순서는 상관엉ㅄ음
plt.show() 
```

## 3번째 미리 다 만들어 놓음 (추천)

```python
fig, axes = plt.subplots(2,2) #fig와 axes객체들이 순서대로 반환됨
axes[0,0].plot(xarr)
axes[1,1].plot(yarr)
plt.show()
```

```python
axes # 4개의 축(2,2)배열
```

```python
fig, axes = plt.subplots(2,2) #fig와 axes객체들이 순서대로 반환됨
axes = axes.flatten() #1차원으로 축소

for i in range(4): # 반복문을 통한 그리기 가능
    axes[i].plot(xarr)
    
plt.show()
```

```python
xarr = np.random.rand(15).cumsum()
yarr = np.random.rand(15).cumsum()
zarr = np.random.rand(15).cumsum()
```

```python
plt.figure(figsize=(6,4))
plt.plot(xarr, label='korean', color='#94B49F', linestyle='--', marker='o') #내부적으로 추가
plt.plot(yarr, label='english', color='#CEE5D0', linestyle='-.', marker='^') # 색상, 레이블 등등 설정이 가능 
plt.plot(zarr, label='japanese', color='#6E85B7', linestyle=':', marker='*')
plt.title('legend', color='#ECB390')
plt.xlabel('years')
plt.ylabel('score')
plt.legend()#범례
plt.show()
#시계열 데이터는 라인플롯으로!
```

## 산점도 scatter plot

```python
plt.scatter(xarr, yarr, color='#000001', s=125)
plt.scatter(xarr, zarr, marker='^')
plt.scatter(yarr, zarr, marker='*')
plt.show()
# x축과 y축의 관계 (상관관계)
```

## Boxplot

```python
#통계 or 데이터 
x = np.random.normal(0,2.0,100)
y = np.random.normal(-3,10.5,100)
z = np.random.normal(1.2,1.5,100)
```

```python
plt.boxplot([x,y,z])
plt.show()
```

* __mean vs redian__
* redian 이상치(이상한 수치)가 있을때 많이 선택함
* 박스의 길이는 데이터가 얼마나 퍼져있는지를 알 수 있음
* 사분위 수 는 이상치를 검거해낼 때 사용
* ex) 미국,한국,일본남성의 평균 몸무게 비교
