```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.__version__
```

## Histogram

```python
titanic = pd.read_csv('./datasets/titanic.csv')
```

```python
titanic.head()
```

```python
titanic.shape
```

```python
type(titanic)
```

```python
plt.figure(figsize=(10,4))
sns.histplot(data=titanic, x='Age')
plt.show() 
```

* metplotlib을 기반으로 만들었기에 plt.show()사용가능
* figsize가 안먹으면 sns자체 함수로 따로 검색해야함

```python
plt.figure(figsize=(10,6))
sns.histplot(data=titanic, x='Age', bins=8)
plt.show() 
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=titanic, x='Age', bins=8)
plt.show() 
```

```python
plt.figure(figsize=(10,4))
sns.histplot(data=titanic, x='Age', kde=True)
plt.show() 
```

```python
plt.figure(figsize=(10,4))
sns.histplot(data=titanic, x='Age', hue='Survived')
plt.show() 
```

* 값에 따라 따로따로 그림을 그리는 것을 요청하는 속성

```python
plt.figure(figsize=(10,4))
sns.histplot(data=titanic, x='Age', bins=8, hue='Sex')
plt.show() 
```

## dispolt
* figure-level plot

```python
sns.displot(data=titanic, x = "Age")
plt.show()
```

```python
sns.displot(data=titanic, x = "Age", kind='kde')
plt.show()
```

```python
sns.displot(data=titanic, x = "Age", col = 'Survived')
plt.show()
```

* displot을 사용하면 두가지 데이터를 쉽게 비교가 가능
* matplotlib에서는 축을 두개 만들고 각각 만들어야 하는 수고를 줄여줌

```python
sns.displot(data=titanic, x = "Age", col = 'Survived', hue = 'Sex')
plt.show()
```

## 막대 그래프 

```python
titanic['Pclass'].value_counts()
```

```python
plt.bar('Pclass', titanic)
```

* matplotlib에서는 위와같이 먼저 나눠주고 막대그래프를 그려야 하지만

```python
sns.countplot(data=titanic, x='Pclass')
```

* countplot은 알아서 데이터를 나누고 계산하여 위와같은 그래프를 그려줌

```python
sns.barplot(data=titanic, x='Pclass', y='Fare')
```

* Pclass별 Fare를 전체중 퍼센테이지로 표현하는 barplot도 가능

```python
sns.barplot(data=titanic, x='Pclass', y='Fare', ci=False, hue='Sex')
```

## Point Plot

```python
sns.pointplot(x='Pclass', y='Fare', hue='Sex', data = titanic)
```

## Boxplot

```python
sns.boxplot(x='Pclass', y = 'Age', data=titanic)
```

## ViolinPlot

```python
sns.violinplot(x='Pclass', y='Age', data = titanic)
```

```python
sns.violinplot(x='Pclass', y='Age', data = titanic)
```

```python
sns.violinplot(x='Pclass', y='Age', data = titanic, split = True, hue='Sex')
```

## Scatter Plot

```python
tips = sns.load_dataset('tips')
type(tips)
```

```python
tips.head()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

* x축과y축 데이터의 관계를 보려는 그래프
* 낸 금액에 따른 tip의 관계

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
sns.regplot(x='total_bill', y='tip', data=tips)
```
